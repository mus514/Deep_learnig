```python
from io import open
import os
from typing import List
import re

from util import process_dir, get_gold, score
```

```python
with open('data/dev/dabo.html', 'r', encoding='ISO-8859-1') as file:
    full_text = file.read()
    print(full_text)
```

```python
example = 'The email address is bouhsen.m@gmail.com'
simple_pattern = r'([\w\.]+)@([\w\.]+)\.([\w\.]{2,})'

matches = re.findall(simple_pattern, example)

m = matches[0]
print('The first group is: %s'%m[0])
print('The second group is: %s'%m[1])
print('The third group is: %s'%m[2])
print("The first and second capture groups are: %s and %s and %s" % m)

email = '%s@%s.%s' % m
print('The extracted email is: %s' % email)
```

```python
def test_find_emails(full_text: str) -> List[str]:
    simple_pattern = r'([\w\.]+)@([\w\.]+)\.([\w]{2,})'
    matches = re.findall(simple_pattern, full_text)

    res = []
    emails = ['%s@%s.%s' % m for m in matches]
    return emails
```

```python
example_line = 'The email address is mustapha.bouhsen@example.edu.'
test_find_emails(example_line)
```

```python
def test_process_file(filename: str, data_directory: str):
    file_name, ext = filename.split('.')
    absolute_path = os.path.join(data_directory, filename)
    res = []

    with open(absolute_path, 'r', encoding='ISO-8859-1') as file:
        full_text = file.read()
        emails = [(file_name, 'e', e) for e in test_find_emails(full_text)]

        res += emails

    return res
```

```python
test_process_file('dabo.html', 'data/dev')
```
